#Importing the required libraries

```python
import pandas as pd
import numpy as np
```

#Reading the csv file

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

##Determining the data types

```python
df.dtypes
```

## 

#using drop to delete the column Unnamed: 0

```python
df=df.drop(columns=["Unnamed: 0"])
df.columns
```

## we can see that the column Unnamed: 0 has been removed

```python
df.head()
```

## cleaning the column names using replace to remove % and the spaces.

```python
df.columns=df.columns.str.replace('[%,(,)]', '')
df.columns=df.columns.str.strip()
df.columns
```

## we can see that the column names with % has now been change to just names without the %

##Determinig the data type of activity_day

```python
df.activity_day.dtypes
```

##changing the object type into DateTime

```python
df.activity_day=pd.to_datetime(df.activity_day)
```

## the data type of activity_day has been changed to DateTime

```python
df.activity_day.dtypes
```

#using fillna to replace NaN with 0 in total_steps as 0 is an appropriate value there as most of the activity type that have NaN in it does not have number of steps involded in it.

```python
df.total_steps=df.total_steps.fillna(0)
df.total_steps=df.total_steps.astype(int)
df.head(10)
```

#splitting the string from ":" to seperate hour and minute and convertin avg_pace to float.

```python
date_format= df[df.avg_pace.notnull()]["avg_pace"].str.split(':')
hour,min = date_format.str[0], date_format.str[1]  

df.loc[df.avg_pace.notnull(),"avg_pace"]=hour.astype(float) + (min.astype(float) / 60) 
```

```python
type(df.avg_pace[3])
```

##Grouping data by workout_type and find the mean, median, count and standard deviation of calories

```python
df.groupby("workout_type")["calories"].agg({"mean","median","count","std"})
```

# initially assuming all values as False and setting up condition for True

```python
df["30%_more"]=0
df.loc[df.aerobic>=30,"30%_more"]=1
df
```

##finding which day of the week has maximum cadence

```python
df["day_of_week"]=df.activity_day.dt.day_name()
df.groupby("day_of_week")["max_cadence"].max()
```

##finding which workout type has maximum cadence

```python
df.groupby("workout_type")["max_cadence"].max()
```

#combining the day of the week, workout type and max cadence 

```python
df[df.max_cadence==df.max_cadence.max()][["day_of_week","workout_type","max_cadence"]]
```
