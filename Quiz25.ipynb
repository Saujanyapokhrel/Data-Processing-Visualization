```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
```

```python
df.plot.hist()
```

```python
plt=df.price.plot.hist(alpha = 1,title = "price histogram", bins=100, xlim=[0,1000], edgecolor='#ffffff')
plt.set_xlabel("Price")
```

```python
plt1=df.groupby('neighbourhood_group').price.plot.hist(alpha = 1, title = "Price by Neighbourhood!", bins=100, xlim=[0, 1000], legend =True)
```
