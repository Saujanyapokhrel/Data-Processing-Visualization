```python
%%bigquery --project data-visualization-338722
SELECT station_id, property_type,
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
limit 50
```

```python
%%bigquery --project data-visualization-338722
WITH t as(
SELECT name, property_type
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
)
SELECT t.property_type as starting_station_type, 
    AVG (trips.duration_minutes) as average_min,
    COUNT(*) as number_of_trips,
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips
  JOIN t
  ON trips.start_station_name = t.name

GROUP BY t.property_type
ORDER by average_min
```

Combining the property_type as starting_station_type and computing the number of trips and the average time in minutes as average_min from austin_bikeshare.bikeshare_stations and displaying the information as one table.
