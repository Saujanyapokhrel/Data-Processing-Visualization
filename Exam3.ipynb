#Histogram of goals

```python
ax = df.Goals.hist(bins = 15)
ax.set(title = "Histogram of Goals")
ax.set_xlabel("Number of Goals", labelpad=10, size=10)
ax.set_ylabel("Number of Players", labelpad=10, size=10)
plt.show()
```

#a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

```python

df[['Goals','Assists']].plot.hist(alpha = 0.8, bins = 15)
```

#a barchart for most goals scored by nationality

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'bar')
ax.set(title = "Most goals scored by nationality")
ax.set(ylabel = 'Number of goals')
plt.show()
```

##from the barchat we can see that players from Egypt scored the most number of goals.

#a boxplot of ‘Yel’ (yellow cards) by nationality.

```python

bp = df.boxplot(column=['Yel'], by = 'nationality', rot = 50)
bp.plot()
plt.xlabel('nationality')
plt.ylabel('yellow cards')
```

#‘Goals’ and ‘Assists’ as scatter signifying Manchester United.

```python
sns.relplot(data=df, x = 'Goals', y= 'Assists' )
plt.title('A scatterplot of goals and assists of all the players')
```

#stacked barplot of goals and assists of French National Team

```python
df3 = df[df['nationality'].str.contains('France')]
ax = df3[['Goals', 'Assists']].plot(kind = 'bar', stacked = True,figsize=(25, 15) , title = 'Stacked barplot of golas and assist of French National Team')
ax.set_xlabel('Goals')
ax.set_ylabel('Assists')
```

Assists are also one of the main attribute to determine the player rating so it should ne stacked on goals as number of goals determine which team wins or loses a match.

#a correlation heatmap of the entire dataset.

```python
sns.set(rc = {'figure.figsize':(30,10)})
ax = sns.heatmap(df.corr(),vmin = -1, annot=False)
ax.set(title = 'Heatmap of World Cup Player Dataset')
plt.show()
```
