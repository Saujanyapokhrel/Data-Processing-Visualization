```python
%%bigquery --project data-visualization-338722
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
```

```python
%%bigquery --project data-visualization-338722
SELECT COUNT(*) as number_of_late_night_false_alarms
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
WHERE incident_group = "False Alarm" AND (hour_of_call >= 22 OR hour_of_call <= 5)
```

there are 3098 false alarms between 10pm  to 5 am.

```python
%%bigquery --project data-visualization-338722
SELECT incident_group, AVG(num_pumps_attending) as avg_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

getting the average numbernum_pumps_attending at the time of incident.
